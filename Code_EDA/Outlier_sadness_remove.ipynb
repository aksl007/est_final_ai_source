#### 수정된 데이터 확인

```python
import pandas as pd
import numpy as np
```

```python
df_train_sadness = pd.read_csv('/workspace/#Ub370#Uc774#Ud130/Data/label/train/train_sadness_1.csv')
df_val_sadness = pd.read_csv ('/workspace/#Ub370#Uc774#Ud130/Data/label/val/val_sadness.csv')

npz_train_sadness = np.load('/workspace/#Ub370#Uc774#Ud130/Data/segmentation/train/train_sadness.npz')
npz_val_sadness = np.load('/workspace/#Ub370#Uc774#Ud130/Data/segmentation/val/val_sadness.npz')
```

```python
# csv 파일 내 이미지 이름들과 npz 파일 내 이미지 이름들이 동일한지 확인 (train_sadness)

csv_filenames = df_train_sadness['filename'].tolist()
npz_filenames = list(npz_train_sadness.keys())

# 동일 여부 확인
common_filenames = set(csv_filenames) & set(npz_filenames)  # 교집합
only_in_json = set(csv_filenames) - set(npz_filenames)      # JSON에만 있는 파일
only_in_npz = set(npz_filenames) - set(csv_filenames)       # NPZ에만 있는 파일

# 결과 출력
print(f"공통 파일 개수: {len(common_filenames)}")
print(f"JSON에만 있는 파일 개수: {len(only_in_json)}")
print(f"NPZ에만 있는 파일 개수: {len(only_in_npz)}")
```

```python
# csv 파일 내 이미지 이름들과 npz 파일 내 이미지 이름들이 동일한지 확인 (val_sadness)

csv_filenames = df_val_sadness['filename'].tolist()
npz_filenames = list(npz_val_sadness.keys())

# 동일 여부 확인
common_filenames = set(csv_filenames) & set(npz_filenames)  # 교집합
only_in_json = set(csv_filenames) - set(npz_filenames)      # JSON에만 있는 파일
only_in_npz = set(npz_filenames) - set(csv_filenames)       # NPZ에만 있는 파일

# 결과 출력
print(f"공통 파일 개수: {len(common_filenames)}")
print(f"JSON에만 있는 파일 개수: {len(only_in_json)}")
print(f"NPZ에만 있는 파일 개수: {len(only_in_npz)}")
```
