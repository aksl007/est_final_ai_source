```python
import pandas as pd

```

```python
album_anger = pd.read_csv('/workspace/Data_Augumentation/Albumentation/label/train/albumentation_train_anger.csv')
album_happy = pd.read_csv('/workspace/Data_Augumentation/Albumentation/label/train/albumentation_train_happy.csv')
album_panic = pd.read_csv('/workspace/Data_Augumentation/Albumentation/label/train/albumentation_train_panic.csv')
album_sadness = pd.read_csv('/workspace/Data_Augumentation/Albumentation/label/train/albumentation_train_sadness.csv')
```

```python
album_anger['path'] = '/workspace/Data_Augumentation/Albumentation/image/train/anger/'
album_happy['path'] = '/workspace/Data_Augumentation/Albumentation/image/train/happy/'
album_panic['path'] = '/workspace/Data_Augumentation/Albumentation/image/train/panic/'
album_sadness['path'] = '/workspace/Data_Augumentation/Albumentation/image/train/sadness/'
```

```python
# 세로로 결합 (행 기준)
df = pd.concat([album_anger, album_happy, album_panic, album_sadness], ignore_index=True)

df.head()
```
