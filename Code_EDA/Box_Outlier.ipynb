```python
df_train_anger = pd.read_csv(json_path_train + 'train_anger.csv')
df_val_anger = pd.read_csv (json_path_val + 'val_anger.csv')

df_train_happy = pd.read_csv(json_path_train + 'train_happy.csv')
df_val_happy = pd.read_csv (json_path_val + 'val_happy.csv')

df_train_panic = pd.read_csv(json_path_train + 'train_panic.csv')
df_val_panic = pd.read_csv (json_path_val + 'val_panic.csv')

df_train_sadness = pd.read_csv(json_path_train + 'train_sadness.csv')
df_val_sadness = pd.read_csv (json_path_val + 'val_sadness.csv')
```

```python
# 이미지 크기와 바운딩 박스 비교
def check_bounding_boxes_with_coords(df, image_folder):
    results = []

    for _, row in df.iterrows():      
        file_name = row['filename']     
        img_path = os.path.join(image_folder, file_name)

        # OpenCV로 이미지 로드
        image = cv2.imread(img_path)
        if image is None:
            print(f"Error loading {img_path}")
            continue

        # 이미지 크기 가져오기
        img_height, img_width, _ = image.shape

        # 각 바운딩 박스(A, B, C, avg)의 좌표 리스트
        x_coords = [row['annot_A.boxes.minX'], row['annot_B.boxes.minX'], row['annot_C.boxes.minX'], row['avg_box.minX']]
        y_coords = [row['annot_A.boxes.minY'], row['annot_B.boxes.minY'], row['annot_C.boxes.minY'], row['avg_box.minY']]
        x_max_coords = [row['annot_A.boxes.maxX'], row['annot_B.boxes.maxX'], row['annot_C.boxes.maxX'], row['avg_box.maxX']]
        y_max_coords = [row['annot_A.boxes.maxY'], row['annot_B.boxes.maxY'], row['annot_C.boxes.maxY'], row['avg_box.maxY']]

        # 박스별로 경계 확인
        for i, box_type in enumerate(['A', 'B', 'C', 'avg']):
            x_min, y_min = x_coords[i], y_coords[i]
            x_max, y_max = x_max_coords[i], y_max_coords[i]

            # 경계를 벗어나는지 확인
            if x_min < 0 or y_min < 0 or x_max > img_width or y_max > img_height:
                results.append({
                    'filename': row['filename'],
                    'box_type': box_type,
                    'x_min': x_min,
                    'y_min': y_min,
                    'x_max': x_max,
                    'y_max': y_max,
                    'image_width': img_width,
                    'image_height': img_height,
                })

    return pd.DataFrame(results)
```

```python
# Check
bounding_box_issues_train_anger = check_bounding_boxes_with_coords(df_train_anger, img_path_train_anger)
bounding_box_issues_val_anger = check_bounding_boxes_with_coords(df_val_anger, img_path_val_anger)

bounding_box_issues_train_happy = check_bounding_boxes_with_coords(df_train_happy, img_path_train_happy)
bounding_box_issues_val_happy = check_bounding_boxes_with_coords(df_val_happy, img_path_val_happy)

bounding_box_issues_train_panic = check_bounding_boxes_with_coords(df_train_panic, img_path_train_panic)
bounding_box_issues_val_panic = check_bounding_boxes_with_coords(df_val_panic, img_path_val_panic)

bounding_box_issues_train_sadness = check_bounding_boxes_with_coords(df_train_sadness, img_path_train_sadness)
bounding_box_issues_val_sadness = check_bounding_boxes_with_coords(df_val_sadness, img_path_val_sadness)
```
