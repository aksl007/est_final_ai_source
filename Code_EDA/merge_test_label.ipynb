```python
import json
import pandas as pd
```

```python
test_path = '/workspace/#Ub370#Uc774#Ud130/test/'
test_json_path = '/workspace/#Ub370#Uc774#Ud130/test/label/'

img_path_test_anger = '/workspace/#Ub370#Uc774#Ud130/test/image/anger/'
img_path_test_happy = '/workspace/#Ub370#Uc774#Ud130/test/image/happy/'
img_path_test_panic = '/workspace/#Ub370#Uc774#Ud130/test/image/panic/'
img_path_test_sadness = '/workspace/#Ub370#Uc774#Ud130/test/image/sadness/'
```

```python
# json 파일 각각 불러오기
with open (test_json_path + "test_anger.json", "r", encoding='euc-kr') as f:
    json_test_anger = json.load(f)
with open (test_json_path + "test_happy.json", "r", encoding='euc-kr') as f:
    json_test_happy = json.load(f)
with open (test_json_path + "test_panic.json", "r", encoding='euc-kr') as f:
    json_test_panic = json.load(f)
with open (test_json_path + "test_sadness.json", "r", encoding='euc-kr') as f:
    json_test_sadness = json.load(f)
```

```python
df_test_anger = pd.json_normalize(json_test_anger)
df_test_happy = pd.json_normalize(json_test_happy)
df_test_panic = pd.json_normalize(json_test_panic)
df_test_sadness = pd.json_normalize(json_test_sadness)

```

```python
df_test_anger['path'] = img_path_test_anger
df_test_happy['path'] = img_path_test_happy
df_test_panic['path'] = img_path_test_panic
df_test_sadness['path'] = img_path_test_sadness
```

```python
df_test = pd.concat([df_test_anger, df_test_happy, df_test_panic, df_test_sadness], axis=0)
```

```python
df_test.info()
```

```python
# 평균 열 생성
df_test['maxX'] = df_test[['annot_A.boxes.maxX', 'annot_B.boxes.maxX', 'annot_C.boxes.maxX']].mean(axis=1)
df_test['maxY'] = df_test[['annot_A.boxes.maxY', 'annot_B.boxes.maxY', 'annot_C.boxes.maxY']].mean(axis=1)
df_test['minX'] = df_test[['annot_A.boxes.minX', 'annot_B.boxes.minX', 'annot_C.boxes.minX']].mean(axis=1)
df_test['minY'] = df_test[['annot_A.boxes.minY', 'annot_B.boxes.minY', 'annot_C.boxes.minY']].mean(axis=1)
```
